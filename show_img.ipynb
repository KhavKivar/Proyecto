```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
from PIL import Image
img = Image.open('dog.jpg').convert('L')
img.save('dog_grey.jpg')
imgConvert = np.array(img)
print(imgConvert.shape)


np.savetxt("dog.dat",imgConvert,fmt='%i')

```

```python
img_bw = np.loadtxt('dog.dat') 
plt.figure(figsize=(10,10))
plt.imshow(img_bw,cmap='gray')
```

```python
foutput = 'dog_gpu_output.dat'
out = np.loadtxt(foutput)
plt.figure(figsize=(10,10))
plt.imshow(out,cmap='gray')

out = out/np.max(out)
thres = 0.065
out1 = np.where(out > thres, out ,1)
plt.figure(figsize=(10,10))
plt.imshow(out1,cmap='gray')
```
